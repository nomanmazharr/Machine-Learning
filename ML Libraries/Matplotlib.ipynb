```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [14,10,8], linewidth= 2)  # this takes data in it which we want to plot
plt.title('Simple chart') # will gives us heading of the chart
plt.ylabel("This is y-axis")
plt.xlabel("This is x-axis")
plt.xscale("linear") # tells us which kind of chart we are plotting, Has attributes like log, logit, function, symlog etc
```
