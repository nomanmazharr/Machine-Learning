```python
import pandas as pd
import seaborn as sn
```

```python
data = sn.load_dataset('tips')
data.head()
```

## Corelation using seaborn heatmap:

```python
corr = data.corr()
```

```python
corr
```

```python
sn.heatmap(corr,annot=True)
```

## Jointplot:
- It is used to plot graph between two axis, like 2D.

```python
sn.jointplot(x='tip',y='size',data=data,kind='')
```

## Pairplot:
- It is used to get graphical representation of whole data

```python
sn.pairplot(data,hue='day') # hue is used to divide on base of that specific column or value that we have given
```

## Distplot:
- It works exactly like hist in matplotlib

```python
sn.displot(data['tip'])
```

### Above all were int or float data types plot functions,


### For categorical data types we use these seaborn graphs types these can also be use for int or float datatypes.
- countplot
- boxplot
- barplot
- violinplot

```python
sn.countplot(x = 'smoker',data = data)
```

```python
sn.violinplot(x = 'tip',y='day',data = data) # one(x,y) should be integer form
```

```python
sn.boxplot(x='sex',y='total_bill',data=data) # also one (s,y) should be integer
```

```python
sn.barplot(x='time',y='tip',data=data)
```
