## Read data using API

```python
url_api = 'https://api.themoviedb.org/3/tv/popular?api_key=44170ecca5ea8893c0f53b779b8819a6&language=en-US&page=1'
req_api = requests.get(url_api)
pd.DataFrame(req_api.json()['results'])
```

```python
# movies = pd.DataFrame()
movies = pd.DataFrame()
for i in range(1,501):
    request_api = requests.get( 'https://api.themoviedb.org/3/tv/popular?api_key=44170ecca5ea8893c0f53b779b8819a6&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame((request_api.json()['results']))
    movies = movies.append(temp_df,ignore_index=True)
```

```python
movies
```

```python
movies.to_csv('movies.csv')
```

### Read data using web scrapping

```python
url_scrapping='https://en.wikipedia.org/wiki/List_of_companies_of_Pakistan'
req = requests.get(url_scrapping,'lxml')
```

```python
soup = BeautifulSoup(req.text)
```

```python
soup.prettify()
```

```python
table = soup.find('table',class_="wikitable sortable")
```

```python
tds = soup.table.find_all('td')
```

```python
for a in tds:
    print(a.text.strip())
```
