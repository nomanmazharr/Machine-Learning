```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
req = requests.get(url)
comp = req.text
```

```python
comp
```

```python
soup = BeautifulSoup(comp,'lxml')
```

```python
table = soup.find('table',class_='wikitable sortable')
```

```python
headers = []
for i in table.find_all('th',scope='col'):
    headers.append(i.text.strip())
```

```python
headers
```

```python
df = pd.DataFrame(columns=headers)
```

```python
table = soup.find('table',{'class':'wikitable sortable'})
table.find_all('th',scope='col')
```

```python
company_new = []
revenue_new = []
profit_new = []
employees_new = []
headquarters = []
name_of_comp = []

for r in table.find_all('tr')[1:]:
    r1 = r.find_all('td')
    if len(r1)>1:
        company_link = r1[1].find('a')
        company = company_link.text if company_link else r1[1].text
        revenue = r1[2].text
        profit = r1[3].text
        employees = r1[4].text
        hq = r1[5].text
        comp_name = r1[0].text
        
        company_new.append(company.strip())
        revenue_new.append(revenue.strip())
        profit_new.append(profit.strip())
        employees_new.append(employees.strip())
        headquarters.append(hq.strip())
        name_of_comp.append(comp_name.strip())
# r1
```

```python
len(r1)
```

```python
df['Name'] = name_of_comp
df['Industry'] = company_new
df['Revenue'] = revenue_new
df['Profit'] = profit_new
df['Headquarters[note 1]'] = headquarters
df['Employees'] = employees_new
```

```python
df
```

```python
df.drop(['Rank','State-owned','Ref.','USD millions'],axis=1,inplace=True)
```
